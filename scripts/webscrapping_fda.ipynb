```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as BS
import re
```

```python
URL = 'https://www.fda.gov/food/outbreaks-foodborne-illness/investigations-foodborne-illness-outbreaks'
response = requests.get(URL)
soup = BS(response.text)
print(soup.prettify())
```

```python
tds = soup.findAll('td')
```

```python
td_text = []
for td in tds:
    x = (td.text.strip())
    td_text.append(x)
td_text
```

```python
td_text = [re.sub('\n\t\t\t', ' ', i) for i in td_text]
```

```python
td_text
```

```python
d = re.compile('\d+\/\d+\/\d{4}')
dates = list(filter(d.match, td_text))
print(dates)
```

```python
bacteria = td_text.copy()
```

```python
bacteria.remove('3/29/2023')
bacteria.remove('1141')
```

```python
bacteria[::11]
```

```python
source = bacteria.copy()
```

```python
source.pop(0)
```

```python
source[::11]
```

```python
FDA_core = pd.DataFrame(list(zip(dates, bacteria[::11], source[::11])),
               columns =['date', 'bacteria', 'source'])
```
