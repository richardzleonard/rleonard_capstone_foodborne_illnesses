```python
import pandas as pd
import re
import numpy as np
import warnings
warnings.filterwarnings('ignore')
from collections import Counter
```

```python
who_AMRO = pd.read_csv('../data/who_AMRO.csv')
```

```python
who_AMRO.sort_values(by = 'Year', ascending = False)
```

```python
canada_who = who_AMRO[['ContaminationIndividualID', 
                       'FoodName', 
                       'FoodGroupName', 
                       'ContaminantName', 
                       'Year']].loc[who_AMRO['CountryName'] == 'Canada']
```

```python
us_who = who_AMRO[['ContaminationIndividualID', 
                       'FoodName', 
                       'FoodGroupName', 
                       'ContaminantName', 
                       'Year']].loc[who_AMRO['CountryName'] == 'United States of America']
```

```python
who_AMRO.groupby('CountryName').count()
```

```python
canada_who = canada_who.drop(['FoodName'], 
                             axis = 1).rename(columns = {'ContaminationIndividualID': 'count'}).groupby(['FoodGroupName', 
                                                                                                         'Year', 
                                                                                                         'ContaminantName']).count()
```

```python
us_who = us_who.drop(['FoodName'], 
                     axis = 1).rename(columns = {'ContaminationIndividualID': 'count'}).groupby(['FoodGroupName', 
                                                                                                 'Year', 
                                                                                                 'ContaminantName']).count()
```

```python
canada_who.to_csv('../data/canada_who.csv')
```

```python
us_who.to_csv('../data/us_who.csv')
```

```python
canada_us_covid = pd.read_csv('../data/canada_us_covid.csv')
```

```python
canada_us_covid = canada_us_covid[['location', 'date', 'population']]
```

```python
canada_us_covid
```

```python
canada_pop = canada_us_covid.loc[canada_us_covid.location == 'Canada']
```

```python
us_pop = canada_us_covid.loc[canada_us_covid.location == 'United States']
```

```python
canada_pop['year'] = canada_pop['date'].str[:4]
```

```python
canada_pop = canada_pop.groupby('year').mean()
```

```python
us_pop['year'] = us_pop['date'].str[:4]
```

```python
us_pop = us_pop.groupby('year').mean()
```

```python
canada_pop.to_csv('../data/canada_pop.csv')
```

```python
us_pop.to_csv('../data/us_pop.csv')
```
